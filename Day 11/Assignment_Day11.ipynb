```python
##import os
import pandas as pd
from scipy.stats import pearsonr

dataset = pd.read_csv("/Users/mypc/Desktop/LetsUpgrade/CaseStudy_Attrition/attrition.csv")
dataset.head()
```

```python
dataset.duplicated()
```

```python
dataset.drop_duplicates()
```

```python
## Converting the attrition values into 0 and 1

dataset['Attrition'] = dataset['Attrition'].replace(['Yes','No'],[1, 0])
dataset.head()
```

### Hypothesis Formulation: 1
__Ho:__ There is no significant relation between Attrition and Age of employees.<br/>
__H1:__ There is significant relation between Attrition and Age of employees.

```python
##Find the correlation between Attrition and Age of employees

a1 = dataset.Attrition
a2 = dataset.Age
stat, p = pearsonr(a1,a2)
print(stat, p)
```

Since p value is less than 0.05, Ho is rejected and H1 is accepted.<br/>
__Inference from the above analysis:__<br/>
There is significant relation between Attrition and Age of employees
